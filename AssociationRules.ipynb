```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
# use Seaborn styles
sns.set()
```

```python
data = pd.read_csv('/Users/Gassantos/DATASET_ITENSFREQUENTES_REGRASASSOCIACAO.csv', index_col=0)
print(data)
```

```python
# getting an overview of our data
print("Our data has {0} rows and {1} columns".format(data.shape[0], data.shape[1]))
# checking for missing values
print("Are there missing values? {}".format(data.isnull().any().any()))
data.describe()
```

```python
colunas = ['IDALUNO', 'PRIMEIROSEMESTRE', 'STATUSFORMACAO', 'DISCIPLINA', 'ANODESVINCULACAO']
pd.pivot_table(data, index=colunas, values="NOTA")
```

```python
pd.pivot_table(data, index='IDALUNO', columns='DISCIPLINA', values="NOTA")
```

```python
d = data.reset_index()
{k: v.tolist() for k, v in d.groupby("IDALUNO")["NOTA"]}
#{1: [9, 28, 40, 47, 51, 89, 93, 99], 2: [5, 11, 44, 85, 93]}
```

```python
pd.pivot_table(data, index='DISCIPLINA', values="NOTA").plot(kind='bar')
plt.ylabel("NOTA")
```

```python
pd.pivot_table(data, index=['DISCIPLINA', 'STATUSFORMACAO'], values="NOTA", aggfunc= [np.mean, np.median, np.std])
```
